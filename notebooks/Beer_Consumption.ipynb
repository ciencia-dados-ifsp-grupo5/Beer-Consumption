# 4. Modelo de regressão linear

````python
from collections import namedtuple

'''
Creates a named tuple to be used as a result from the ``run_model`` function (https://docs.python.org/3/library/collections.html#collections.namedtuple).
The attributes are as follow:
    name : str
        Semantic name of the model.
    features : list
        List of features names used to train the the model.
    params : ndarray
        Array of regression parameters: ``np.append(model.intercept_, model.coef_)``.
    r2_train : float
        Coefficient of determination R^2 of the trained model: ``model.score(X_train, y_train)``.
    r2_test : float
        R^2 (coefficient of determination) regression score function from sklearn.metrics: ``metrics.r2_score(y_test, y_pred)``.
    mae : float
        Mean Absolute Error regression loss: ``metrics.mean_absolute_error(y_test, y_pred)``.
    mse : float
        Mean Squared Error regression loss: ``metrics.mean_squared_error(y_test, y_pred)``.
    rmse : float
        Root Mean Squared Error regression loss: ``metrics.mean_squared_error(y_test, y_pred, squared=False)``.
    y_test : array
        Ground truth (correct) target values used to estimations test.
    y_pred : array
        Estimated target values.
    model : instance of sklearn.linear_model.LinearRegression
        Instance of the fitted linear regression model
'''
ModelResults = namedtuple('ModelResults', ['name', 'features', 'params', 'r2_train', 'r2_test', 'mae', 'mse', 'rmse', 'y_test', 'y_pred', 'model'])

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Função para criar modelo
def run_model(df, name, features=None, target='consumo_de_cerveja_litros', train_size=0.8, random_state=1):
    '''
    Instantiate, train and make predictions using a Linear Regression Model.
    
    Parameters
    ----------
    df : DataFrame
        Dataset to be used in training and testing the model.
    name : str
        Semantic name to the model.
    features : list, optional, default ``None``
        List of features to be used as independent variables columns.
    target : str, optional, default ``consumo_de_cerveja_litros``
        Column name of the target variable.
    train_size : float or int, optional, default ``0.8``
        Passed directly as param to sklearn ``train_test_split``.
        If float, should be between 0.0 and 1.0 and represent the
        proportion of the dataset to include in the train split. If
        int, represents the absolute number of train samples.
    random_state : int, RandomState instance or None, default ``1``
        Passed directly as param to sklearn ``train_test_split``.
        Controls the shuffling applied to the data before applying the split.
        Pass an int for reproducible output across multiple function calls.
        
    Returns
    -------
    results: ModelResults
        A named tuple ``ModelResults`` object, as defined above.
    '''
    
    # Selecionando atributos
    if features == None:
        features = df.columns[df.columns != target]
        
    X = df[features].copy()
    y = df[target].copy()
    
    # Dividindo base em treino e test: Default: 80% treino e 20% teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, random_state=random_state)
    
    # Criando modelo
    model = LinearRegression().fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    results = ModelResults(
        name = name,
        features = features,
        params = np.append(model.intercept_, model.coef_),
        r2_train = model.score(X_train, y_train),
        r2_test = metrics.r2_score(y_test, y_pred),
        mae = metrics.mean_absolute_error(y_test, y_pred),
        mse = metrics.mean_squared_error(y_test, y_pred),
        rmse = metrics.mean_squared_error(y_test, y_pred, squared=False),
        y_test = y_test,
        y_pred = y_pred,
        model = model
    )
    return results
````

````python
# Versão 1: Todos os atributos
# Versão 2: Somente Temperatura
# Versão 3: Sem precipitação
# Versão 4: Temperatura Máxima e FDS
# Versão 5: Temperatura máxima, precipitação e fds
models_to_test = [
    ('Todos atributos', list(dados.columns[1:-1])),
    ('Somente Temperatura', ['temperatura_media_c', 'temperatura_minima_c', 'temperatura_maxima_c']),
    ('Sem precipitação', ['temperatura_media_c', 'temperatura_minima_c', 'temperatura_maxima_c', 'final_de_semana']),
    ('Temperatura máxima e fds', ['temperatura_maxima_c', 'final_de_semana']),
    ('Temperatura máxima, precipitação e fds', ['temperatura_maxima_c', 'precipitacao_mm', 'final_de_semana'])
]

results = []
for model in models_to_test:
    result = run_model(dados, name=model[0], features=model[1])
    results.append(result)
````

````python
# Parâmetros da regressão
rows = []
for result in results:
    row = pd.DataFrame(columns=['Intercepto'] + result.features)
    row.loc[result.name] = result.params
    rows.append(row)
pd.concat(rows).style.format('{:.2f}').set_caption('<b>Parâmetros dos modelos de regressão linear</b>')
````

````python
# Erros
pd.DataFrame(
    [(r.r2_train, r.r2_test, r.mae, r.mse, r.rmse) for r in results],
    columns=['R² treino', 'R² teste', 'MAE', 'MSE', 'RMSE'],
    index=[r.name for r in results]
).style.format('{:.2f}').set_caption('<b>Métricas de Avaliação dos modelos</b>')
````

````python
fig, axs = plt.subplots(nrows=len(results), ncols=3, sharex='col', figsize=(22, 4*len(results)))

for i, result in enumerate(results):
    
    # Gráfico de Dispersão: Valores reais vs preditos
    ax = sns.scatterplot(x=result.y_pred, y=result.y_test, ax=axs[i][0])
    ax.set(title=f'{result.name}: Previsto X Real', xlabel='Valor Previsto', ylabel='Valor Real')
    ax.xaxis.set_tick_params(which='both', labelbottom=True)
    
    # Histograma: Erro
    ax = sns.histplot(result.y_pred - result.y_test, bins=20, kde=True, ax=axs[i][1])
    ax.set(title=f'Frequência do erro', xlabel='Erro da previsão', ylabel='Frequência absoluta')
    ax.xaxis.set_tick_params(which='both', labelbottom=True)
    
    # Violinplot: Distribuição dos erros
    ax = sns.violinplot(x=abs(result.y_pred - result.y_test), cut=0, ax=axs[i][2])
    ax.set(title=f'Distribuição dos erros absolutos', xlabel='')
    ax.xaxis.set_tick_params(which='both', labelbottom=True)

plt.subplots_adjust(hspace = 0.5 )
fig.suptitle('4.1 - Gráficos dos erros dos modelos', fontsize=22, y=0.94)

plt.show()
````

**Modelo**: O modelo utilizado foi o [LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) do `sklearn.linear_model`. Foi construida uma função para organizar e simplificar a criação de versões do modelo variando os atributos usados. 

**Treino e Teste**: A divisão da base foi feita utilizando o  [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) e como padrão utilizamos 80% para treino e 20% para testes.

**Métricas**: Do pacote [`metrics` do sklearn](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics), utilizamos como parâmetros de avaliação o R², MAE, MSE e RMSE. 

**Atributos**: Utilizamos 5 combinações diferentes de atributos para treinar o modelo
* `Versão 1:` Todos os atributos (temperaturas média, mínima e máxima, precipitação e final de semana)
* `Versão 2:` Somente Temperatura (temperaturas média, mínima e máxima)
* `Versão 3:` Sem precipitação (temperaturas média, mínima e máxima e final de semana)
* `Versão 4:` Temperatura Máxima e fds (temperatura máxima e final de semana)
* `Versão 5:` Temperatura máxima, precipitação e fds (temperatura máxima, precipitação e final de semana)

**Parametros da Regressão**: A **Precipitação** possui um coeficiente negativo, indicando que o aumento desse valor reduz o consumo de cerveja. A **Temperatura Máxima** possui um coeficiente médio de 700, indicando que a cada grau aumentado de temperatura máxima, o consumo de cerveja aumenta em cerca $700$ litros. O atributo que mais teve influência foi **Final de Semana**, com um coeficiente aproximado de 5.100. Os valores dos coeficientes de **Temperatura Média** e **Temperatura Mínima** nos modelos acabam sofrendo a influência dos valores de Temperatura Máxima devido à alta correlação desses atributos, de modo que seus valores pouco interferem no modelo.

**Resultados**: O Melhor valor obtido foi com o modelo da `Versão 5 - Temperatura máxima, precipitação e fds`, obtendo um valor de MAE de $1.922,68$ e RMSE de $2.314,46$. Outros modelos que tiveram um resultado próximo foram a `Versão 1 - Todos os atributos` e a `Versão 4 - Temperatura Máxima e fds`. O pior modelo encontrado foi a `Versão 2 - Somente Temperatura`. Isso pode ter acontecido devido a alta correlação dos valores e a exclusão da variável de final de semana, que se mostrou importante na análise exploratória e na análise dos parametros da regressão.

Considerando que a distribuição da variável alvo no *dataset* em questão é de $25.401 \pm 4.399$ litros, podemos concluir que nossos modelos tiveram um resultado satisfatório. No melhor caso, conseguimos realizar predições com um RMSE de pouco mais de **meio desvio padrão dos dados**. Diversos dos modelos atingiram também um coeficiente de determinação ($R^2$) superiores a $0,7$, indicando que mais de $70\%$ da variação do consumo de cerveja pode ser explicada pelos atributos que consideramos para regressão.

## 4.3. Colocando o modelo em produção

Como última etapa, vamos levar o modelo para produção a partir de um acesso via API.

Para tanto, primeiramente exportamos o modelo com melhor desempenho (no caso, consideramos o menor RMSE). Vamos exportar também a lista de atributos utilizada pelo modelo, de forma que a função na API saiba quais colunas devem estar presentes nos dados de entrada pelo usuário.

````python
import pickle

best_result = min(results, key=lambda m: m.rmse)
print('Melhor modelo:', best_result.name)

best_model = best_result.model
best_model_features = best_result.features

with open('../deploy/backend/models/beer_consumption_model.pkl', 'wb') as model_file:
    pickle.dump(best_model, model_file)
    
with open('../deploy/backend/models/beer_consumption_features.pkl', 'wb') as features_file:
    pickle.dump(best_model_features, features_file)
````

A [API desenvolvida](https://github.com/ciencia-dados-ifsp-grupo5/Beer-Consumption/tree/main/deploy/backend/api) é bastante simples. Basicamente é criada uma aplicação em [Flask](https://flask.palletsprojects.com/en/2.0.x/) que carrega os arquivos do modelo e das colunas exportados e expõe uma única rota em `/model/predict` com método `POST`. O corpo da requisição esperado é em formato JSON e deve conter um único elemento ou uma lista de elementos no formato do exemplo:

```json
{
    "temperatura_media_c": 22.76,
    "temperatura_minima_c": 19.7,
    "temperatura_maxima_c": 28,
    "precipitacao_mm": 85.3,
    "final_de_semana": 0   
}
```

Note que não necessariamente todos atributos enviados pela consulta serão usados pelo modelo. Justamente por isso a API carrega também as colunas com as quais o modelo foi treinado. No nosso caso, por exemplo, o modelo que foi para "produção" não utiliza os atributos de `temperatura_media_c` e `temperatura_minima_c`, ou seja, requisições com quaisquer valores desses atributos terão o mesmo retorno. O retorno será uma lista de valores das predições (consumo de cerveja em litros), com o mesmo número de elementos da lista enviada para consulta.

Por simplicidade, uma vez que nosso objetivo era apenas demostrar o conceito, a aplicação não realiza diversas verificações dos dados e considera que as consultas sejam feitas no formato esperado. Se um atributo esperado pelo modelo não constar na consulta, seu valor será preenchido com $0$.

O endereço da consulta mostrado na imagem (http://54.172.94.89:5000/model/predict) está funcional e pode ser acessado publicamente (mas lembre-se que o método permitido para esta rota é `POST`, portanto não será possível acessar diretamente pelo navegador de internet).
